# AoU Workbench Bootstrap

This notebook clones or pulls `spand67/aou-workbench`, installs it in editable mode, and runs preflight.

```python
from pathlib import Path
import subprocess

REPO_URL = "git@github.com:spand67/aou-workbench.git"
REPO_DIR = Path.home() / "aou-workbench"
if REPO_DIR.exists():
    subprocess.run(["git", "-C", str(REPO_DIR), "pull", "--ff-only", "origin", "main"], check=False)
else:
    subprocess.run(["git", "clone", REPO_URL, str(REPO_DIR)], check=False)
REPO_DIR
```

```python
subprocess.run(["python", "-m", "pip", "install", "--user", "--no-deps", "-e", "."], cwd=REPO_DIR, check=False)
```

```python
subprocess.run(["aou-workbench", "preflight"], cwd=REPO_DIR, check=False)
```
